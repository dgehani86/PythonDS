```python
womandf = pd.read_csv("https://raw.githubusercontent.com/akjadon/HH/master/Python/DS_tutorial_trn/91.Introduction%20to%20Data%20Visualizaion%20with%20Python/percent-bachelors-degrees-women-usa.csv")
womandf.head()
```

```python
year = womandf.Year
ps = womandf['Physical Sciences']
cs = womandf['Computer Science']

plt.plot(year, ps, 'bo-')
```

```python
plt.plot(year, cs, color='red')
```

```python
plt.show
```

```python
# Plot in blue the % of degrees awarded to women in the Physical Sciences using axis
plt.axes([0.05,0.05,0.425,0.9])
plt.plot(year,ps,'b-')
```

```python
# Plot in red the % of degrees awarded to women in Computer Science
plt.axes([0.525,0.05,0.425,0.9])
plt.plot(year, cs, color='red')
plt.show
```

```python
# Plot the % of degrees awarded to women in Computer Science and the Physical Sciences
plt.plot(year,cs, color='red')
plt.plot(year,ps, color='blue')
plt.xlabel('Year')
plt.ylabel('Degree in %')
plt.xlim((1990,2010))  #Using xlim(), ylim()
plt.ylim((0,50))
plt.title('Degree awarded to woman (1990 - 2010)\nComputer Science (red) | Physical Sciences (blue)')
plt.show
#plt.savefig('xlim_and_ylim.png')
```

```python
plt.plot(year,cs, color='red')
plt.plot(year,ps, color='blue')
plt.xlabel('Year')
plt.ylabel('Degree in %')
plt.axis((1990,2010,0,50))
plt.title('Degree awarded to woman (1990 - 2010)\nComputer Science (red) | Physical Sciences (blue)')
plt.show
```

```python
#Using legend()
plt.plot(year,cs, color='red' ,label='Computer Science') 
plt.plot(year,ps, color='blue',label='Physical Sciences')
plt.legend(loc = 'lower center')
plt.xlabel('Year')
plt.ylabel('% Enrollment')
plt.title('Undergrad enrollment of Woman')
```

# -------------------------------------------------------------------------------------------#

```python
cs_max = cs.max()
print(cs_max)
yr_max = womandf.iloc[womandf['Computer Science'].idxmax()].Year
print(yr_max)
```

```python

#Using annotate()
plt.plot(year,cs, color='red' ,label='Computer Science') 
plt.plot(year,ps, color='blue',label='Physical Sciences')
plt.legend(loc = 'lower right')
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+5, cs_max+5), arrowprops=dict(facecolor='black'))
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')
plt.title('Undergraduate enrollment of women')
plt.show()
```

```python
plt.style.use('ggplot')
plt.subplot(2, 2, 1) 
plt.plot(year, ps, color='blue')
plt.title('Physical Sciences')
```

```python
plt.style.use('ggplot')
plt.subplot(2, 2, 1) 
plt.plot(year, ps, color='blue')
plt.title('Physical Sciences')

plt.subplot(2, 2, 2)
plt.plot(year, cs, color='red')
plt.title('Computer Science')

plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max-1, cs_max-10), arrowprops=dict(facecolor='black'))

plt.subplot(2, 2, 3)
plt.plot(year, health, color='green')
plt.title('Health Professions')

# Plot the enrollment % of women in Education
plt.subplot(2, 2, 4)
plt.plot(year, education, 'y-')
plt.title('Education')
plt.tight_layout()
```
